```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/renatomaaliw3/public_files/master/Data%20Sets/Iris%20(5).csv")
df.head()
```

```python
df['Species'].unique()
```

```python
df = df.drop('Id', axis = 1)
```

```python
sns.pairplot(df, hue = 'Species')
```

```python
sns.catplot(x = 'Species', y = 'PetalLengthCm', data = df, kind = 'box', hue = 'Species')
```

```python
from sklearn.model_selection import train_test_split as tss

x = pd.get_dummies(df.drop('Species', axis = 1), drop_first = True)
y = df['Species']

xtrain, xtest, ytrain, ytest = tss(x, y, test_size = 0.3, random_state = 50)
```

```python
#show x
x
```

```python
#show y
y
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
model.fit(xtrain, ytrain)
```

```python
predict = model.predict(xtest)
predict
```

```python
from sklearn.metrics import confusion_matrix, classification_report as cr
```

```python
cm = confusion_matrix(ytest,predict)
sns.heatmap(cm, annot = True, xticklabels = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'], yticklabels = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']).set_title('Confusion Matrix')
```

```python
print(cr(ytest, predict))
```

```python
imp = model.feature_importances_
pd.DataFrame(index = x.columns, columns = ['Feature Importance'], data = imp, ).sort_values('Feature Importance', ascending = False)
```

```python
from sklearn import tree
text_representation = tree.export_text(model, feature_names = x.columns.tolist())
print(text_representation)
```

```python
from sklearn.tree import plot_tree as pt

plt.figure(figsize = (12,8), dpi =  200)
pt(model, feature_names = x.columns, filled = True)
```

```python
#alternate tree
import graphviz

dot_data = tree.export_graphviz(model, feature_names = x.columns, filled = True)

graph = graphviz.Source(dot_data, format = 'png')
graph
```

```python
model2 = DecisionTreeClassifier(max_depth = 3, random_state = 50)
model2.fit(xtrain, ytrain)

plt.figure(figsize = (12,8), dpi =  200)
pt(model2, feature_names = x.columns, filled = True)
```

```python
model3 = DecisionTreeClassifier(max_leaf_nodes = 4, random_state = 50)
model3.fit(xtrain, ytrain)

plt.figure(figsize = (12,8), dpi =  200)
pt(model3, feature_names = x.columns, filled = True)
```

```python
model4 = DecisionTreeClassifier(max_depth = 4, max_leaf_nodes = 3, random_state = 50)
model4.fit(xtrain, ytrain)

plt.figure(figsize = (12,8), dpi =  200)
pt(model4, feature_names = x.columns, filled = True)
```
